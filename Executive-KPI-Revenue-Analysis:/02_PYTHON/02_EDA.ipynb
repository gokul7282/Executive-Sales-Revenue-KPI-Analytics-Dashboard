# Final Insights

1. Technology category generated the highest revenue among all categories.

2. APAC market contributed the highest sales and showed strong business performance.

3. A small group of customers contributed a significant portion of total revenue.

4. Several products generated negative profit despite good sales performance.

5. Discount showed a negative relationship with profit.

6. Revenue consistently increased over the years indicating business growth.

# Business Recommendations

1. Increase marketing investment in Technology products.

2. Review pricing and discount strategy for loss-making products.

3. Focus customer retention programs on high-value customers.

4. Expand operations in high-performing regions such as APAC.

5. Monitor products with low profitability and optimize inventory.
